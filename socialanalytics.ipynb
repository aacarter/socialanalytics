```python
import tweepy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

from config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
target_users = ("@BBCBreaking", "@CBSNews", "@cnnbrk",
                "@FoxNews", "@nytimes")
#oldest_tweet = None
counter = 0

compound_list = []
positive_list = []
negative_list = []
neutral_list = []
date = []
content = []
username = []
tweets_ago = []

for target in target_users:


    for x in range(5):

        public_tweets = api.user_timeline(target, page = x) #max_id=oldest_tweet)

        for tweet in public_tweets:
            if counter % 100 == 0:
                counter = 0
            counter += 1
            
            results = analyzer.polarity_scores(tweet["text"])
            
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
            account = (tweet["user"]["name"])
            copy = (tweet["text"])
            time = (tweet["created_at"])
            
            
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
            content.append(copy)
            date.append(time)
            username.append(account)
            tweets_ago.append(counter)
            #oldest_tweet = (tweet["id"] - 1)
            
                    
        news_mood = pd.DataFrame ({
            "Account": username,
            "Text": content,
            "Date": date,
            "Compound": compound_list,
            "Positive": positive_list,
            "Neutral": negative_list,
            "Negative": neutral_list,
            "Tweets Ago": tweets_ago
            })
```

```python
news_mood.head(25)
```

```python
BBC = news_mood[news_mood["Account"] == "BBC Breaking News"]
CBS = news_mood[news_mood["Account"] == "CBS News"]
CNN = news_mood[news_mood["Account"] == "CNN Breaking News"]
FOX = news_mood[news_mood["Account"] == "Fox News"]
NYT = news_mood[news_mood["Account"] == "The New York Times"]

```

```python
ax = BBC.reset_index().plot(kind="scatter", x="Tweets Ago", y="Compound", color = 'lightskyblue', grid=True, 
                                    edgecolors="black", label = 'BBC', alpha = .75)
CBS.reset_index().plot(kind="scatter", x="Tweets Ago", y="Compound", color = 'green', grid=True, 
                                  ax=ax, edgecolors="black", label = 'CBS', alpha = .75)
CNN.reset_index().plot(kind="scatter", x="Tweets Ago", y="Compound", color = 'red', grid=True, 
                                  ax=ax, edgecolors="black", label = 'CNN', alpha = .75)
FOX.reset_index().plot(kind="scatter", x="Tweets Ago", y="Compound", color = 'blue', grid=True, 
                                  ax=ax, edgecolors="black", label = 'FOX', alpha = .75)
NYT.reset_index().plot(kind="scatter", x="Tweets Ago", y="Compound", color = 'yellow', grid=True, 
                                  ax=ax, edgecolors="black", label = 'NYT', alpha = .75)

plt.title("Sentiment Analysis of Media Tweets (10/4/2018)",fontsize=14)
plt.xlabel("Tweets Ago)", fontsize=14)
plt.ylabel("Tweet Polarity", fontsize=14)

plt.legend(loc= "upper right", bbox_to_anchor=(1.2, 1))
plt.savefig("News_Plot.png")
```
